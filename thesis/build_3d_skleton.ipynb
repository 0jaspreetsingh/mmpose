```python
path = '/home/jaspreet/Downloads/Thesis/code/Sept 2/mmpose/vis_results/extracted_pose_3d.pkl'
updated_path = '/home/jaspreet/Downloads/Thesis/code/Sept 2/mmpose/vis_results/extracted_pose_3d_updated.pkl'
import mmengine
import numpy as np
import os

extracted_pose_3d = mmengine.load(path)
error_videos = extracted_pose_3d['error_causing_videos']

# Iterate over the keys in split and filter out the videos in error_causing_videos
for split_key, video_list in extracted_pose_3d['split'].items():
    extracted_pose_3d['split'][split_key] = [video for video in video_list if video not in error_videos]


mmengine.dump(extracted_pose_3d, updated_path)
```
